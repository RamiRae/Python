```python
df = pd.read_csv('benin-malanville.csv')

# --- Create Polar Plot ---

# Convert wind direction to radians
df['WD_rad'] = np.deg2rad(df['WD'])

# Create the polar plot
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})

# Plot wind speed as radial values and direction as angular values
# Store the scatter plot object in a variable
scatter_plot = ax.scatter(df['WD_rad'], df['WS'], s=20, c=df['WS'], cmap='viridis')

# Set plot labels and title
ax.set_theta_zero_location("N")  # Set North at the top
ax.set_theta_direction(-1)  # Clockwise direction
ax.set_rlabel_position(90)  # Move radial labels to the right
ax.set_title("Wind Speed and Direction Distribution")

# Add a colorbar to show wind speed values
plt.colorbar(mappable=scatter_plot, label='Wind Speed (m/s)')  # Pass the scatter plot object

# --- Calculate Wind Direction Variability ---

# Calculate standard deviation of wind direction (in degrees)
wind_direction_std = df['WD'].std()

# Add text annotation for wind direction variability
ax.text(0.5, 0.9, f"Wind Direction Variability: {wind_direction_std:.1f}",
        transform=ax.transAxes, ha='center', va='top', bbox={'facecolor': 'white', 'alpha': 0.8})

plt.show()
```

```python

df = pd.read_csv('benin-malanville.csv')
null_columns = df.columns[df.isnull().all()]
print(f"Columns with entirely null values: {null_columns}")

# Drop columns with entirely null values
df.drop(columns=null_columns, inplace=True)

# Fill missing values in other columns with appropriate methods
# Example: Fill missing values in 'GHI' with the mean
df['GHI'].fillna(df['GHI'].mean(), inplace=True)
df['GHI_ZScore'] = (df['GHI'] - df['GHI'].mean()) / df['GHI'].std()
threshold = 2
df['GHI_Outlier'] = (abs(df['GHI_ZScore']) > threshold).astype(int)
df['GHI'] = np.where(df['GHI_Outlier'] == 1, df['GHI'].median(), df['GHI'])


# Drop the 'WS_Outlier' column
df['DNI'].fillna(df['DNI'].mean(), inplace=True)
df['DNI_ZScore'] = (df['DNI'] - df['DNI'].mean()) / df['DNI'].std()
threshold = 2
df['DNI_Outlier'] = (abs(df['DNI_ZScore']) > threshold).astype(int)
df['DNI'] = np.where(df['DNI_Outlier'] == 1, df['DNI'].median(), df['DNI'])

# --- Remove Outlier Flags ---
df['DHI'].fillna(df['DHI'].mean(), inplace=True)
df['DHI_ZScore'] = (df['DHI'] - df['DHI'].mean()) / df['DHI'].std()
threshold = 2
df['DHI_Outlier'] = (abs(df['DHI_ZScore']) > threshold).astype(int)
df['DHI'] = np.where(df['DHI_Outlier'] == 1, df['DHI'].median(), df['DHI'])

# --- Remove Outlier Flags ---

# Drop the 'WS_Outlier' column
df.drop(columns='DHI_Outlier', inplace=True)
# --- Remove Z-Score Column ---

# Drop the 'WS_ZScore' column
df.drop(columns='DHI_ZScore', inplace=True)
# --- Remove Outlier Flags ---

# Drop columns with outlier flags
df.drop(columns=[col for col in df.columns if '_Outlier' in col], inplace=True)

# --- Remove Z-Score Columns ---

# Drop columns with Z-scores
df.drop(columns=[col for col in df.columns if '_ZScore' in col], inplace=True)

# --- Save the Cleaned DataFrame ---

df.to_csv('cleaned_data.csv', index=False)
print(df)
```
